## Remove Punctuation

```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    punctuations = string.punctuation
    return text.translate(str.maketrans('','',punctuations))
```

```python
df['clean_text'] = df['content'].apply(lambda x: remove_punctuation(x))
```

## Cleaning Text

```python
import re

def cleaning_text(text):
    text = re.sub(r'[^\u0980-\u09FF\s]', '', str(text))
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_text'] =df['clean_text'].apply(cleaning_text)
```

```python
df
```

## Tokenize

```python
from bnltk.tokenize import Tokenizers
import re

tokenizer = Tokenizers()

def tokenize(text):
      tokens = tokenizer.bn_word_tokenizer(str(text))

      return ' '.join(tokens)


df['clean_text'] =df['clean_text'].apply(lambda x: tokenize(x))
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.10, stratify=df['category'], random_state=42)
train_df, val_df = train_test_split(train_df, test_size=0.10, stratify=train_df['category'], random_state=42)
print(f"Train: {len(train_df)}, Validation: {len(val_df)}, Test: {len(test_df)}")
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

MAX_NUM_WORDS = 50000
tokenizer = Tokenizer(num_words=MAX_NUM_WORDS, oov_token='<OOV>')
tokenizer.fit_on_texts(train_df['clean_text'].astype(str))
vocab = set(list(tokenizer.word_index.keys())[:MAX_NUM_WORDS])
print('Vocabulary size (tokenizer.word_index):', len(tokenizer.word_index))
print('Using vocab subset size:', len(vocab))
```

```python
from pathlib import Path

FT_GZ = 'cc.bn.300.vec.gz'
FT_VEC = 'cc.bn.300.vec'
if not Path(FT_GZ).exists() and not Path(FT_VEC).exists():
    print('Downloading cc.bn.300.vec.gz (this may take a while)...')
    # Use wget which is available on Kaggle; fall back to requests if not.
    try:
        !wget -q -c https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.bn.300.vec.gz -O {FT_GZ}
    except Exception as e:
        print('wget failed, trying requests...')
        import requests
        url = 'https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.bn.300.vec.gz'
        with requests.get(url, stream=True) as r:
            r.raise_for_status()
            with open(FT_GZ, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    if chunk:
                        f.write(chunk)
    print('Download finished.')
else:
    print('FastText archive already present.')
```

```python
import gzip
from tqdm import tqdm

filtered_vec = 'cc.bn.300.filtered.vec'
if not Path(filtered_vec).exists():
    print('Filtering FastText to only keep words present in tokenizer vocab...')
    need = set(vocab)  # words we need
    found = 0
    # The .vec.gz file has a header on the first line like: <num_words> <dim>
    with gzip.open(FT_GZ, 'rt', encoding='utf-8', errors='ignore') as zin, open(filtered_vec, 'w', encoding='utf-8') as fout:
        header = next(zin)  # skip header; we'll write a placeholder later
        # Temporarily store matches to update header counts later
        fout.write('0 300\n')  # placeholder -- overwritten later
        for line in tqdm(zin, desc='scanning vectors'):
            parts = line.rstrip().split(' ')
            if not parts:
                continue
            word = parts[0]
            if word in need:
                fout.write(line)
                found += 1
    # Update header with correct count
    with open(filtered_vec, 'r+', encoding='utf-8') as f:
        clean_text = f.read()
        f.seek(0,0)
        f.write(str(found) + ' 300\n' + clean_text.split('\n',1)[1])
    print(f'Filtered vectors saved to {filtered_vec}. Words found: {found}')
else:
    print('Filtered FastText vectors already exist.')
```

```python
import numpy as np
embedding_index = {}
filtered_vec = 'cc.bn.300.filtered.vec'
with open(filtered_vec, 'r', encoding='utf-8', errors='ignore') as f:
    header = f.readline()
    for line in f:
        values = line.rstrip().split(' ')
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs
print('Loaded embeddings for', len(embedding_index), 'words.')
```

```python
import numpy as np

embedding_dim = 300
num_words = min(MAX_NUM_WORDS, len(tokenizer.word_index)) + 1
embedding_matrix = np.zeros((num_words, embedding_dim), dtype='float32')

missing = 0
for word, i in tokenizer.word_index.items():
    if i >= num_words:
        continue
    vec = embedding_index.get(word)
    if vec is not None:
        embedding_matrix[i] = vec
    else:
        missing += 1
print('Embedding matrix shape:', embedding_matrix.shape)
print('Missing word vectors:', missing)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences

MAX_LEN = 100
X_train_seq = pad_sequences(tokenizer.texts_to_sequences(train_df['clean_text'].astype(str)), maxlen=MAX_LEN)
X_val_seq = pad_sequences(tokenizer.texts_to_sequences(val_df['clean_text'].astype(str)), maxlen=MAX_LEN)
X_test_seq = pad_sequences(tokenizer.texts_to_sequences(test_df['clean_text'].astype(str)), maxlen=MAX_LEN)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(train_df['category'])
y_val = le.transform(val_df['category'])
y_test = le.transform(test_df['category'])
num_classes = len(le.classes_)
print('Num classes:', num_classes)
```

#ML AND DL


```python
import tensorflow as tf
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional, Conv1D, GlobalMaxPooling1D

tfidf = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf.fit_transform(train_df['clean_text'].astype(str))
X_val_tfidf = tfidf.transform(val_df['clean_text'].astype(str))
X_test_tfidf = tfidf.transform(test_df['clean_text'].astype(str))

ml_models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'SVM': LinearSVC(),
    'Random Forest': RandomForestClassifier(n_estimators=100, n_jobs=-1)
}

ml_results = {}
for name, model in ml_models.items():
    print('\nTraining', name)
    model.fit(X_train_tfidf, y_train)
    preds = model.predict(X_test_tfidf)
    acc = accuracy_score(y_test, preds)
    ml_results[name] = acc
    print(f'{name} test accuracy: {acc:.4f}')
    print(classification_report(y_test, preds, target_names=le.classes_))
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=69)

model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
EMBEDDING_DIM = embedding_dim
model_bilstm = Sequential([
    Embedding(input_dim=num_words, output_dim=EMBEDDING_DIM, weights=[embedding_matrix],
              input_length=MAX_LEN, trainable=False),
    Bidirectional(LSTM(64, dropout=0.2, recurrent_dropout=0.2)),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])
model_bilstm.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_bilstm.summary()

history_bilstm = model_bilstm.fit(X_train_seq, y_train, validation_data=(X_val_seq, y_val),
                                  epochs=5, batch_size=32)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history_bilstm.history['loss'], label='train_loss')
plt.plot(history_bilstm.history['val_loss'], label='val_loss')
plt.legend(); plt.title('BiLSTM Loss')

plt.subplot(1,2,2)
plt.plot(history_bilstm.history['accuracy'], label='train_acc')
plt.plot(history_bilstm.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title('BiLSTM Accuracy')
plt.show()
```

```python
model_cnn = Sequential([
    Embedding(input_dim=num_words, output_dim=EMBEDDING_DIM, weights=[embedding_matrix],
              input_length=MAX_LEN, trainable=False),
    Conv1D(128, 5, activation='relu'),
    GlobalMaxPooling1D(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(num_classes, activation='softmax')
])
model_cnn.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model_cnn.summary()

history_cnn = model_cnn.fit(X_train_seq, y_train, validation_data=(X_val_seq, y_val),
                            epochs=5, batch_size=32)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history_cnn.history['loss'], label='train_loss')
plt.plot(history_cnn.history['val_loss'], label='val_loss')
plt.legend(); plt.title('CNN Loss')

plt.subplot(1,2,2)
plt.plot(history_cnn.history['accuracy'], label='train_acc')
plt.plot(history_cnn.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title('CNN Accuracy')
plt.show()
```
